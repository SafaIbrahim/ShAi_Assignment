```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
# Load your dataset
df =pd.read_csv('Salaries.csv')
df.head()
```

# Basic Data Exploration

```python
df.columns
num_rows, num_columns = df.shape
print(f"Number of rows: {num_rows}")
print(f"Number of columns: {num_columns}")
print("\nData types of each column:")
print(df.dtypes)
```

```python
missing_values = df.isnull().sum()
print("\nMissing values in each column:")
print(missing_values)
```

# Descriptive Statistics

```python
df.describe()
```

```python
# Calculate mean
mean_salary = df['TotalPay'].mean()

# Calculate median
median_salary = df['TotalPay'].median()

# Calculate mode
mode_salary = df['TotalPay'].mode()

print(f"Mean Salary: {mean_salary}")
print(f"Median Salary: {median_salary}")
print(f"Mode Salary: {mode_salary}")
```

```python
Max_salary = df['TotalPay'].max()
print ("\nMax salary:",Max_salary)

Min_salary = df['TotalPay'].min()
print ('\nMin Salary:', Min_salary)
```

```python
salary_range = (df['TotalPay'].max() - df['TotalPay'].min())
print("Range of salaries:", salary_range)
```

```python
std_salary = df['TotalPay'].std()
print ("\nstandard deviation:", std_salary)
```

# Data Cleaning

```python
#drop Notes and Status because they both null
columns_to_drop = ['Notes', 'Status']


##The missing values in the 'BasePay', 'Benefits', 'OtherPay', and 'OvertimePay' columns were imputed with the mean of each
#respective column.
#This imputation strategy was chosen under the assumption that the missing values are missing at random.
#The goal was to maintain the overall distribution of salaries and associated benefits while mitigating the impact
#of missing data.


df['BasePay'].fillna(df['BasePay'].mean(), inplace=True)
df['Benefits'].fillna(df['Benefits'].mean(), inplace=True)
df['OtherPay'].fillna(df['OtherPay'].mean(), inplace=True)
df['OvertimePay'].fillna(df['OvertimePay'].mean(), inplace=True)
df.head()
```

# Basic Data Visualization

```python
# Visualization 1: Histogram for Salary Distribution
plt.figure(figsize=(10, 6))
plt.hist(df['TotalPay'], bins=20, color='skyblue', edgecolor='black')
plt.title('Distribution of Salaries')
plt.xlabel('TotalPay')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
import seaborn as sns
# Use pie charts to represent the proportion of employees in different departments
department_counts = df['JobTitle'].value_counts().head(12)
#used 12 out of 2159 Job tittles

plt.pie(department_counts, labels=department_counts.index, autopct='%1.1f%%', startangle=60)
plt.title('Proportion of Employees in Top 12 Departments')
plt.show()
```

# Grouped Analysis

```python
grouped_data = df.groupby(['JobTitle', 'Year'])
summary_statistics = grouped_data['TotalPay'].agg(['count', 'mean', 'median', 'min', 'max'])
print(summary_statistics)
```

# Simple Correlation Analysis

```python
correlation = df['TotalPay'].corr(df['OvertimePay'])
print(f"Correlation between TotalPay and OvertimePay: {correlation}")
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='OvertimePay', y='TotalPay', data=df, alpha=0.5)
plt.title('Scatter Plot: TotalPay vs. OvertimePay')
plt.xlabel('OvertimePay')
plt.ylabel('TotalPay (Salary)')
plt.show()
```
